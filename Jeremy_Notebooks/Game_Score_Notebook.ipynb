```python
def game_standing_score(game_name, date_time):
    desired_row = game_standing_df[(game_standing_df['game_name'] == game_name) &
                     (game_standing_df['timestamp'] == date_time)]
    return float(desired_row['game_standing_ratio'].values[0])
```

```python
def game_growth_score(game_name, date_time):
    desired_row = game_growth_df[(game_growth_df['game_name'] == game_name) &
                 (game_growth_df['timestamp'] == date_time)]
    if np.isnan(desired_row['game_growth'].values[0]):
        print('Game growth score data not available. Returned 1')
        return 1
    else:
        growth_score = desired_row['game_growth'].values[0]
    return float(growth_score)
```

```python
def top_8_market_share_ratio(game_name, date_time):
    date_time = pd.Timestamp(date_time)
    day_of_week = date_time.day_name()
    hour_of_day = date_time.hour
    market_share_value = top8_marketshare_dict[day_of_week,hour_of_day,game_name]
    return float(market_share_value)
```

```python
def calc_game_score(game_name, date_time):
    '''date_time is a timestamp object '''
    game_standing_constant = 1
    game_growth_constant = 1
    market_share_constant = 1
    growth_score = game_growth_score(game_name, date_time) 
    try:
        market_share = top_8_market_share_ratio(game_name,date_time)
    except KeyError:
        market_share = .2
    game_standing = game_standing_score(game_name, date_time)
    game_score = game_growth_constant * growth_score\
                    + market_share_constant * market_share\
                    + game_standing_constant * game_standing
    return game_score
```
