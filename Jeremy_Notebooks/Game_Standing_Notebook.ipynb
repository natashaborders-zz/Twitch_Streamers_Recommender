# Calculate Game Standing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import inf
pd.options.display.max_rows=500
```

```python
colnames=['timestamp', 'game_name', 'num_channels', 'num_viewers']
```

```python
df=pd.read_csv('Natasha.csv',header =None, names = colnames)
```

```python
df=df.sort_values(by=['timestamp']).reset_index(drop=True)
```

```python
df.groupby(by=['timestamp','game_name']).max().groupby(['timestamp']).max()
```

```python
df_max = df.groupby(by='timestamp').max().sort_values(by=['timestamp','num_viewers'], ascending = False)
```

# Max Viewers in Any Sully Hour Period

```python
df_max_per_hour=df[df.groupby(['timestamp'])['num_viewers'].transform(max) == df['num_viewers']]
```

```python
df_max_per_hour.head()
```

```python
game_standing_df = pd.merge(df, df_max_per_hour, on='timestamp', suffixes=['', '_max'])
game_standing_df['game_standing_ratio'] = game_standing_df['num_viewers'] / game_standing_df['num_viewers_max']
```

```python
game_standing_df.to_pickle('game_score_dataframes/game_standing_df.pkl')
```
